<h2> Weather station to county mapping data has 63 Null values in the county name and county columns in 112 values. Match the unknown county values to their closest counties based on longitute and latitude </h2>

```python
location_county_no_nulls = weather_station_to_county_mapping_df.dropna(how='any')
location_county_no_nulls
```

```python
labeled_locations = list(location_county_no_nulls.itertuples(index=False))
labeled_locations
```

```python
  weather_station_to_county_mapping_df[weather_station_to_county_mapping_df.county.isnull()]
```

```python
location_county_nulls = weather_station_to_county_mapping_df[weather_station_to_county_mapping_df.county.isnull()][['longitude', 'latitude']]
location_county_nulls
```

```python
unlabeled_locations = list(location_county_nulls.itertuples(index=False))
unlabeled_locations
```

```python
def find_closest_label(location, labeled_locations):
    closest_label = None
    min_distance = float('inf')

    for labeled_location in labeled_locations:
        distance = geodesic(location, (labeled_location.latitude, labeled_location.longitude)).kilometers
        if distance < min_distance:
            min_distance = distance
            closest_county = labeled_location.county
            closest_county_name = labeled_location.county_name

    return closest_county, closest_county_name
```

```python
assigned_counties = []

for location in unlabeled_locations:
    county, county_name = find_closest_label((location.latitude, location.longitude), labeled_locations)
    assigned_counties.append((county_name, location.latitude, location.longitude, county))
    print(f"Location {location} is closest to {county_name}, county number {county}")
```

```python
assigned_counties_df = pd.DataFrame(assigned_counties, columns = ['county_name', 'longitude', 'latitude', 'county'])
assigned_counties_df
```

```python
# create a new dataframe using imputed county name / county if values
weather_station_to_county_mapping_df_imputed = pd.concat([location_county_no_nulls, assigned_counties_df])
weather_station_to_county_mapping_df_imputed
```
